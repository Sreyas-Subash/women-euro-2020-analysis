```python
import pathlib
import os
import pandas as pd
import json

pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
```

```python
#path to data
path = os.path.join(str(pathlib.Path().resolve()), 'Wyscout', 'competitions.json') 

#open data
with open(path) as f:
    data = json.load(f)
    #save it in dataframe
df_competitions = pd.DataFrame(data)
#structure of data
df_competitions.info()
```
